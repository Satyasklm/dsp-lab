# Visualization 1

```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35,50,15])
plt.pie(y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35,50,15])
co=["r","b","hotpink"]
plt.pie(y,colors=co,labels=["DSP","CD","WT"])
plt.pie(y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35,50,15])
co=["r","b","hotpink"]
plt.pie(y,colors=co,labels=["DSP","CD","WT"])
plt.pie(y)
plt.legend(title="subjects")
plt.show()
```

```python
import matplotlib.pyplot as plt
country=['USA','India','Germany','UK','France']
GDP_per_capita=[48000,38000,50000,45000,44000]
plt.bar(country,GDP_per_capita)
plt.title('country vs GDP_per_capita')
plt.ylabel('GDP per capita')
plt.show()
```

```python
import matplotlib.pyplot as plt
numbers=[0.1,0.5,1,1.5,2,4,5.5,6,8,9]
plt.hist(numbers,bins=3,color="hotpink")
plt.xlabel('Number')
plt.ylabel('Frequency')
plt.show()
```
