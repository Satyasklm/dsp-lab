# Visualization

```python
import matplotlib.pyplot as plt
import numpy as np
plt.title("Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
xpoints=np.array([0,20])
ypoints=np.array([0,100])
plt.plot(xpoints,ypoints)
plt.show()
```

```python
plt.plot([5,100],[5,200],"--")
plt.show()
```

```python
x=[1,2,3,4,5]
y=[10,20,30,40,50]
plt.scatter(x,y)
plt.show()
```

```python
#Multiple points
xpoints=np.array([1,2,6,8])
ypoints=np.array([3,8,1,10])
plt.plot(xpoints,ypoints)
plt.show()
```

```python
#plotting without x-points
ypoints=np.array([3,8,1,10,5,7])
plt.plot(ypoints)
plt.show()
```

```python
#plotting without y-points:
ypoints=np.array([3,8,1,10,5,7])
plt.plot(ypoints,marker="*",mec="r")
plt.show()
```

```python
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,marker="o",mec="b",color="r")
plt.show()
```

```python
#plotting withour x-points:
ypoints=np.array([3,8,1,10,5,7])
plt.plot(ypoints,marker="*",mec="r",linewidth="25")
plt.show()
```
