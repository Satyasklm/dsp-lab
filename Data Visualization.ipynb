```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35,50,15])
plt.pie(y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35,50,15])
co=["r","b","hotpink"]
plt.pie(y,colors=co,labels=["DSP","CD","WT"])
plt.pie(y)
plt.legend(title="subjects")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
country=['USA','India','Germany','UK','France']
GDP_per_capita=[48000,38000,50000,45000,44000]
plt.bar(country,GDP_per_capita)
plt.title('country vs GDP_per_capita')
plt.xlabel("Country")
plt.ylabel('GDP per capita')
plt.show()
```

```python
import matplotlib.pyplot as plt
country=['USA','India','Germany','UK','France']
GDP_per_capita=[48000,38000,50000,45000,44000]
plt.barh(country,GDP_per_capita,color="pink")
plt.title("Bar Graph")
plt.xlabel("Country")
plt.ylabel("Capita")
plt.show()
```

```python
import matplotlib.pyplot as plt
numbers=[0.1,0.5,1,1.5,2,4,5.5,6,8,9]
plt.hist(numbers)
plt.xlabel('Number')
plt.ylabel('Frequency')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.hist(numbers,bins=3,color="pink")
plt.xlabel('Number')
plt.ylabel('Frequency')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([0,1,2,4])
y=np.array([3,8,1,10])
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title("Linee")
x=np.array([0,1,2,3])
y=np.array([10,20,30,40])
plt.subplot(1,2,2)
plt.pie(y)
plt.title("pieee")
plt.suptitle("Graphs")
plt.show()
```
